## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

I'm picking the CTR / Engagement Opportunity Scoring lane. Some pages already rank well in Google (page 1, positions 1–20) and get real search visibility, but they still capture far fewer clicks than other pages sitting at the same position — meaning the traffic is already there, someone just isn't clicking the search result. That's a fast, concrete fix: an editor can rewrite a title or meta description today, without waiting weeks for new content or a ranking change to pay off. It also plugs directly into signals already in the starter dataset (impressions, position, CTR), so I can start testing it immediately.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

For an SEO/content editor with limited time, deciding which already-page-1 pages to rewrite (title, meta description, or snippet) first, I'll build a ranked score from the starter dataset (impressions, average position, position tier, CTR) that flags pages whose CTR sits clearly below other pages at the same position tier. A wrong call has a real cost either way: flagging a page that wasn't actually underperforming wastes editor hours that could have gone to a real opportunity, while missing a genuinely underperforming page means it keeps losing clicks silently, with nobody ever reviewing it. A plain fixed rule like "CTR below 0.5%" isn't enough on its own, because what counts as low CTR depends heavily on position — a rule has to account for that before it's trustworthy, which is the kind of adjustment a simple model or careful segmented analysis is good at.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/ShabbirHussainCodes/flyrank-ml-internship"
REPO_DIR = "flyrank-ml-internship"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)

import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# avg_position == 0 means "no position data" in this dataset, not literally rank zero
has_position = df[df["avg_position"] > 0]

# Number 1: how many "visible but underperforming CTR" pages exist right now
low_ctr_visible = has_position[
    (has_position["impressions_90d"] >= 500) &
    (has_position["avg_position"] <= 20) &
    (has_position["ctr"] < 0.5)
]
print(f"Pages with real visibility (>=500 impressions, position 1-20) but CTR < 0.5%: "
      f"{len(low_ctr_visible)} of {len(has_position)} visible pages")

# Number 2: CTR varies a lot even within the same position tier -> position alone doesn't explain it
ctr_by_tier = has_position.groupby("position_tier")["ctr"].agg(["mean", "std", "count"]).round(3)
print("\nAverage CTR by position tier:")
print(ctr_by_tier)

# Number 3: how many impressions are "stuck" behind these underperforming pages
print(f"\nTotal impressions sitting behind low-CTR visible pages: "
      f"{int(low_ctr_visible['impressions_90d'].sum()):,}")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

This lane can only support observed and directional claims: measured differences in click-through rate between pages that already sit at the same search position, in this anonymized slice of FlyRank's client data. The output is decision-support — a ranked list an editor could choose to review — not a guarantee. It cannot claim that rewriting a title or meta description will fix the gap; that would require an actual before/after experiment, which this data doesn't provide. It says nothing about Google's ranking algorithm, and it is not evidence of AI search visibility or AI citations. A low-CTR flag is a candidate worth a human look, not proof that the title or meta is the actual problem.
